## <font color ='indianred'>**1.1 Change  the hashtags to lower case** 
e.g. replace #Coronavirus with #coronavirus

Create a new column clean_text. The clean_text should have modified hashtags.

```python
search_hashtags = re.compile(r"(?:#+[\w]+[\w'-]*[\w]+)") # regex pattern to find hashtag
df['clean_text'] = df['Tweet Content'].swifter.apply(lambda x: re.sub(search_hashtags,x.lower(), x))
df['clean_text'].swifter.apply(lambda x: re.findall(search_hashtags, x)) # to check if the code worked
```

## <font color ='indianred'>**1.2 Remove RT from text in column clean_text**


```python
# We're creating copies of df to work with
df_2 = df
```

### <font color ='skyblue'> **a. So we are assuming that we need to remove  the entire retweet given that this is already present in the dataset so we are removing duplicates by dropping all rows that start with RT**

```python
# Create a function to make things easier
def search_RT(search_tag, text):
  match = re.findall(search_tag, text)
  if match:
    return True
  return False
```

```python
search_retweets = re.compile(r"^RT\s+")
df_2['RT'] = df_2['clean_text'].swifter.apply(lambda x: search_RT(search_retweets, x))
df_2 = df_2[df_2['RT']==False]; df_2.drop(['RT'], axis=1,inplace=True)
```

```python
# check to see if the desired result is achieved
df_2['clean_text']
```

### <font color ='skyblue'> **b. For this part we are simply removing RT from the tweet content**

```python
retweet_pattern = re.compile(r"\bRT\s+")
df['clean_text'] = df['clean_text'].swifter.apply(lambda x: re.sub(retweet_pattern,'', x))
df['clean_text'] # to check if the code worked
```

## <font color ='indianred'>**1.3 Remove URLs and links from text in column clean_text**

```python
search_url = re.compile(r"((http|ftp|https):\/\/)?[\w\-_]+(\.[\w\-_]+)+([\w\-\.,@?^=%&amp;:/~\+#]*[\w\-\@?^=%&amp;/~\+#])?")
df['clean_text'] = df['clean_text'].swifter.apply(lambda x: re.sub(search_url,'',x))
```

## <font color ='indianred'>**1.4 Removing Punctuations from text in column clean_text.**


```python
df['clean_text']=df['clean_text'].str.translate(str.maketrans('','', string.punctuation))
df['clean_text']
```

## <font color ='indianred'>**1.5 Extract number of Hashtags in a new column num_hashtags**


```python
search_hashtags = re.compile(r"(?:#+[\w]+[\w'-]*[\w]+)") # regex pattern to find hashtag
df['num_hashtags'] = df['Tweet Content'].swifter.apply(lambda x: len(re.findall(search_hashtags, x)))
df['num_hashtags']
```

## <font color ='indianred'>**1.6 Extract number of user mentions in a new column num_mentions**

```python
search_mentions = re.compile(r"(?<=^|(?<=[^a-zA-Z0-9-_\.]))@([A-Za-z0-9_]{4,15})")
df['num_mentions'] = df['Tweet Content'].swifter.apply(lambda x: len(re.findall(search_mentions, x)))
df['num_mentions']
```

## <font color ='indianred'>**Task 2.1 Keep rows with name of  image files only**
We ony need file names for images. The images are in the folder oxford-iiit-pet/images and end with .jpg.

For this task you will need to write a regular expression that matches the file names that have the word `images` in it and end with `.jpg`.

Use this expression to keep ony those rows that corresponds to images.

```python
def drop_rows(search_pattern, text):
  match = re.findall(search_pattern, text)
  if match:
    return True
  return False
```

```python
jpg_pattern = re.compile(".jpg$")
df['check'] = df['filename'].swifter.apply(lambda x: drop_rows(jpg_pattern, x))
df = df[df['check']==True]; df = df.drop(['check'], axis=1,inplace=False)
```

## <font color ='indianred'>**Task 2.2 Extract image names from files names**

The part after `images` in file names represent the image names. For example  for the file name `oxford-iiit-pet/images/newfoundland_61.jpg`, image name is `newfoundland_61.jpg`

Extract only image names from filenames and store them in a separate column.



```python
def extract_image_name(search_pattern, text):
  match = re.search(search_pattern, text)
  if match:
    return match.group(1)
  return None
```

```python
search_filename = re.compile(r"([^/]+$)")
df['image_names'] = df['filename'].swifter.apply(lambda x: extract_image_name(search_filename, x))
```

## <font color ='indianred'>**Task 2.3  Extract label names from image names**
Each image correspond to image of a unique breed of a dog or a cat. The ML task would be to predict the breed of cat/dog given the image file. To do this task we first need to get the outcome (label or dependent varible) from the file names. In this dataset the part before the _number.jpg represents the breed name. For example for image name `german_shorthaired_151.jpg` the breed name is erman_shorthaired. Keep in mind that same breed names are single words and some are multiple words separated by underscores.

Extract the label names as new column in teh dataframe.

```python
def extract_label(x):
  result = re.match(r'^(\w+)_\d+\.jpg', x)
  if result is not None:
    label = result.group(1)
  else:
    label = 'Error' 
  return label
```

```python
df['label'] = df['image_names'].swifter.apply(lambda x: extract_label(x))
```

## <font color ='indianred'>**Task 2.4 Check whether breed name is a Cat or Dog**
If the first alphabet  of the breed name is a capital letter then it is a cat else it is a dog. For example `Siamese` is a Cat whereas `chihuahua` is a dog.

Add a new column `Cat' which takes the value of $1$ if breed corresponds t0 a cat and $0$ if it corresponds to a dog.

```python
def tagging_animal(search_pattern, text):
  result = re.match(search_pattern, text)
  if result is not None:
    return 1
  return 0
```

```python
cat_pattern = re.compile(r"\b[A-Z]")
df['Cat'] = df['label'].swifter.apply(lambda x: tagging_animal(cat_pattern, x))
```
